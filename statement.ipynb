```python
from math import sqrt
def calculate_std(value: list):
    N = len(value)
    mean = sum (value)/ N
    temp = sum([(x - mean) ** 2 for x in value])

    std = sqrt(temp / N)

    return round (std,2)
    
```

```python
calculate_std([8,9,6,4,3,4,5,6,7])
```

```python
class StudentResult:
    def __init__(self,name:str,marks:list[int]):
        self.name = name 
        self.marks = marks

    def total(self):
        return sum(self.marks)

    def average(self):
        return self.total()/len(self.marks)

    def display(self):
        print (f"Name: {self.name}")
        print ("-" * 50)
        print (f"Total marks obtained: {self.total()}")
        print (f"Average Marks: {self.average()}")
        
```

```python
rohit= StudentResult("Rohit",marks=[91,46,13,75])
```

```python
rohit.display()
```

```python
class person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

class employee(person):
    def __init__(self, name: str, age: int, employee_id: int):
        super().__init__(name, age)
        self.employee_id = employee_id

class manager(employee):
    def __init__(self, name: str, age: int, employee_id: int, department="not_defined"):
        super().__init__(name, age, employee_id)
        self.department = department

    def get_details(self):
        print(self.name, self.age, self.employee_id, self.department)
```

```python
m = manager("shirish",23,101)
```

```python
m.get_details()
```

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def __str__(self):
        return f"{self.year} {self.make} {self.model}"

my_car = Car("Toyota", "Camry", 2023)
print(my_car)  # Calls __str__ and prints "2023 Toyota Camry"
print(str(my_car)) # Calls __str__ and prints "2023 Toyota Camry"
```

```python
class point:
    def __init__(self,x,y):
       self.x = x
       self.y = y

    def __str__(self):
        return f"Point({self.x},{self.y})"

    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y

        return point(x,y)
```

```python
A = point(3,4)
B = point(4,7)
```
