```python
from math import sqrt
def calculate_std(value: list):
    N = len(value)
    mean = sum (value)/ N
    temp = sum([(x - mean) ** 2 for x in value])

    std = sqrt(temp / N)

    return round (std,2)
    
```

```python
calculate_std([8,9,6,4,3,4,5,6,7])
```

```python
class StudentResult:
    def __init__(self,name:str,marks:list[int]):
        self.name = name 
        self.marks = marks

    def total(self):
        return sum(self.marks)

    def average(self):
        return self.total()/len(self.marks)

    def display(self):
        print (f"Name: {self.name}")
        print ("-" * 50)
        print (f"Total marks obtained: {self.total()}")
        print (f"Average Marks: {self.average()}")
        
```

```python
rohit= StudentResult("Rohit",marks=[91,46,13,75])
```

```python
rohit.display()
```
